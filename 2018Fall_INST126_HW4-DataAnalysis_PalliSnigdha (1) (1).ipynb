### Question 4 (8 pts total): Male and female income by occupation. 

The dataset "bls.csv" contains the median weekly earnings of full-time wage and salary workers by detailed occupation and sex. The data is structured as comma-separated values (CSV). Each row has seven fields:

- Occupation: Job title as given from BLS. Industry summaries are given in ALL CAPS.
- All_workers: Number of workers male and female, in thousands.
- All_weekly: Median weekly income including male and female workers, in USD.
- M_workers: Number of male workers, in thousands.
- M_weekly: Median weekly income for male workers, in USD.
- F_workers: Number of female workers, in thousands.
- F_weekly: Median weekly income for female workers, in USD.

#### Part 1 (2 pts): Create a new dataframe that contains only the ***industry*** data (i.e., excluding the rows that are about occupations) and assign it to the variable `industries`.

***Notes:***
- In the Occupation column, ***industries*** are designated by upper case letters. ***Occupations*** (types of jobs) are mixed case. Use this distinction to help you answer subquestion a.
- You may find the `df.loc` function helpful
</pre>


```python
# Your code here (you should end up with 23 rows in your dataframe)
bls_df = pd.read_csv("bls.csv")
industries = bls_df.loc[bls_df['Occupation'] == bls_df['Occupation'].str.upper()]
print(industries)
```

#### Part 2 (6 pts): Calculate wage statistics

***In separate cells below, calculate and print the indicated values (1 point each):***<pre>
a. Using the `industries` dataframe, find the industries with the most and least *female* workers

```python
# 2.a. Your code goes here
most_female = industries.loc[industries['F_workers'].idxmax()]
least_female = industries.loc[industries['F_workers'].idxmin()]
print(most_female)
print(least_female)
```

```python
# 2.b. Your code goes here
most_male = industries.loc[industries['M_workers'].idxmax()]
least_male = industries.loc[industries['M_workers'].idxmin()]

print(most_male)
print(least_male)
```

```python
# 2.c. Your code goes here
industries.query('F_workers < 200')
```

```python
# 2.d. Your code goes here
industries['F_weekly'] = industries['F_weekly'].astype(str).astype(float)
industries['M_weekly'] = industries['M_weekly'].astype(str).astype(float)

difference = industries['M_weekly'].mean() - industries['F_weekly'].mean() 

print(difference)
```

```python
# 2.e. Your code goes here
industries['weekly_difference'] = industries['M_weekly'] - industries['F_weekly']
difference_max = industries['weekly_difference'].max()
print(difference_max)
final_max = industries.loc[industries['weekly_difference'] == difference_max]
print(final_max)
```

```python
# 2.f. Your code goes here
industries['weekly_difference'] = industries['M_weekly'] - industries['F_weekly']
difference_min = industries['weekly_difference'].min()
print(difference_min)
final_min = industries.loc[industries['weekly_difference'] == difference_min]
print(final_min)
```
