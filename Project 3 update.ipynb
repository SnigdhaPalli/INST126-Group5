```python
#import matplotlib
import matplotlib.pyplot as plt
%pylab inline

#import pandas
import pandas

# open and read the playsotre csv
df = pandas.read_csv('googleplaystore.csv')
```

```python
# order the data by highest to lowest rating
df_rating_initial = df.sort_values (by = 'Rating', ascending = False)
#drop the first row of initial rating dataset because information was input incorectly
df_rating = df_rating_initial.drop(df_rating_initial.index[0])
# print the new dataset df_rating
print(df_rating)
```

```python
# order the data by highest to lowest rating
df_price_initial = df.sort_values (by = 'Price', ascending = False)
#drop the first row of initial price dataset because information was input incorectly
df_price = df_price_initial.drop(df_price_initial.index[0])
#print the new dataset df_price
print(df_price)
```

```python
#Create a function graphs that takes two colums, titles for each, and a title for the graph
def graphs(x_col, y_col, x_axis, y_axis, title):
    #create a plot using the inputed columns and show the values as only points 
    plt.plot(x_col,y_col,'ro')
    #create a label for the x-axis
    plt.xlabel(x_axis)
    #create a label for the y-axis
    plt.ylabel(y_axis)
    #create a label for the title
    plt.title(title)
    #print the graph
    print(plt)
#Graph of price vs rating
price_g = graphs(df_rating['Price'], df_rating['Rating'], "Price of app", "Rating", "Price vs Rating")
```

```python
# graph of price vs number of installs
graphs(df_rating['Price'], df_rating['Installs'], "Price of app", "Number of Installs", "Price vs Number of Installs")
```
