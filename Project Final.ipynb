```python
#import matplotlib; Diego (from update#2)
import matplotlib.pyplot as plt
%pylab inline

#import pandas; Diego (from update#2)
import pandas as pd 

# open and read the playsotre csv; Diego (from update#2)
df = pd.read_csv('googleplaystore.csv')
```

```python
# order the data by highest to lowest rating; Daniel
df_rating_initial = df.sort_values (by = 'Rating', ascending = False)
#drop the first row of initial rating dataset because information was input incorectly; Daniel
df_rating = df_rating_initial.drop(df_rating_initial.index[0])
# print the new dataset df_rating; Daniel
print(df_rating)
```

```python
# Create a function to delete the $ from an inputted column and return all the data in that column as a float; Snigdha, Daniel
def no_sign(x):
    x = x.replace('$', '')
    return float(x)

# Create a function to delete the + and , from an inputted column and return all the data in that column as a float; Snigdha, Daniel
def no_sign_installs(x):
    x = x.replace('+', '')
    x = x.replace(',', '')
    return float(x)

# Create a column 'price_num' with the data from the column 'Price,' which has been refined by the no_sign function; Snigdha
df_rating['price_num'] = df_rating['Price'].apply(no_sign)
#print price_num; Snigdha
print(df_rating["price_num"])

```

```python
# order the data by highest to lowest price; Daniel
df_price_initial = df.sort_values (by = 'price_num', ascending = False)
#drop the first row of initial price dataset because information was input incorectly; Daniel
df_price = df_price_initial.drop(df_price_initial.index[0])

#print the new dataset df_price; Daniel
print(df_price)
```

```python
#Create a function graphs that takes two colums, titles for each, and a title for the graph; Snigdha
def graphs(x_col, y_col, x_axis, y_axis, title):
    #create a plot using the inputed columns and show the values as only points 
    plt.plot(x_col,y_col,'ro')
    #create a label for the x-axis
    plt.xlabel(x_axis)
    #create a label for the y-axis
    plt.ylabel(y_axis)
    #create a label for the title
    plt.title(title)
    #print the graph
    plt.xticks(rotation = 90)
    print(plt)

#Graph of price vs rating; Snigdha
price_g = graphs(df_rating['price_num'], df_rating['Rating'], "Price of app", "Rating", "Price vs Rating")
```

```python
#covert the Installs column from a string; Snigdha 
df_rating['installs_num'] = df_rating['Installs'].apply(no_sign_installs)

#graph of price vs number of installs; Snigdha 
graphs(df_rating['price_num'], df_rating['installs_num'], "Price of app", "Number of Installs", "Price vs Number of Installs")
```
